# Relationship of the data features with the target

```python
data["target_names"]
```

```python
df["target_names"] = df["target"].map({0: "setosa", 1: "versicolor", 2: "virginica"})
```

```python
col = "sepal length (cm)"
sns.relplot(x=col, y="target", hue="target_names", data=df)
plt.suptitle(col, y=1.05)
plt.show()
```

```python
col = "sepal width (cm)"
sns.relplot(x=col, y="target", hue="target_names", data=df)
plt.suptitle(col, y=1.05)
plt.show()
```

```python
col = "petal length (cm)"
sns.relplot(x=col, y="target", hue="target_names", data=df)
plt.suptitle(col, y=1.05)
plt.show()
```

```python
col = "petal width (cm)"
sns.relplot(x=col, y="target", hue="target_names", data=df)
plt.suptitle(col, y=1.05)
plt.show()
```
